```python
#Number guessing game
import random as r
import math as m
lower_number = int(input("Enter the lower number: "))
upper_number = int(input("Enter the upper number: "))
random_number = r.randint(lower_number, upper_number)
total_chances=m.ceil(m.log(upper_number - lower_number + 1, 2))
print("total chances you have:",total_chances)
count=0
f=False
for(count) in range(total_chances):
    guess=int(input("Guess a number: "))

    if random_number==guess:
        print("Congratulations you did it in", count+1, "try")
    print("Done", count)
    if count==total_chances-1:
          f=True
          break
    elif random_number>guess:
        print("You guessed too small")
    elif random_number<guess:
        print("You guessed too high")

if not f:
          print("The number is", random_number)
          print("Better luck next time")
```
